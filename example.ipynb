```python
%%sql

SELECT *
FROM news
LIMIT 1
```

```python
%%sql

SELECT "public"."news"."containsWordGlobalWarming" AS "containsWordGlobalWarming", date_trunc('month', "public"."news"."date") AS "date", count(*) AS "count"
FROM "public"."news"
GROUP BY "public"."news"."containsWordGlobalWarming", date_trunc('month', "public"."news"."date")
ORDER BY "public"."news"."containsWordGlobalWarming" ASC, date_trunc('month', "public"."news"."date") ASC
```

```python
%%sql

SELECT title, url, "public"."news"."date"
FROM "public"."news"
WHERE ((lower("public"."news"."description") like '%changement climatique%')
   OR "public"."news"."containsWordGlobalWarming" = TRUE)
ORDER BY date_trunc('month', "public"."news"."date") ASC
```
