```python
import pandas as pd
import json
from glob import glob
import os.path as osp
import datetime
import wordcloud
from io import StringIO
import stop_words
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
import re
```

```python
pd.set_option('display.max_colwidth', 500)
```

```python
media = 'France 2'
edition = 'jt-de-20h'
```

```python
subjects = []
day_files = glob(f'../data-news-json/media={media}/year=2022/month=*/day=*/*.json')
for day_file in day_files:    
    for line in open(day_file):
        subject = json.loads(line)
        subject['media'] = media.lower().replace(' ', '')
        subjects.append(subject)
```

```python
data = pd.DataFrame(subjects)
```

```python
data = data[[edition in x for x in data.urlTvNews]]
```

```python
data[['date', 'title', 'description', 'url']].sort_values('date').head()
```
