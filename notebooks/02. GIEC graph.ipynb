```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
%matplotlib inline
```

```python
def getImage(path, zoom=1):
   return OffsetImage(plt.imread(path, format="jpg"), zoom=zoom)
```

```python
data = pd.read_csv('/home/david/Documents/giec.csv', encoding="utf16")
```

```python
data
```

```python
global_count = data.drop('Chaine', axis=1).sum(axis=0)
```

```python
global_count.plot(kind='bar')
```

```python
ax = sns.barplot(y=global_count, x=global_count.index)
ab = AnnotationBbox(getImage('/home/david/Downloads/logo-climatmedias-cropped-scaled.png', zoom=1), (1.7,8000), frameon=False)
ax.add_artist(ab)
```

```python
focus = 'France info'
```

```python
focus_count = data.set_index('Chaine').loc[focus]
ax = sns.barplot(y=focus_count, x=focus_count.index)
ax.set_title(focus)
ax.set_ylabel('')
ab = AnnotationBbox(getImage('/home/david/Downloads/logo-climatmedias-cropped-scaled.png', zoom=1), (1.8,2500), frameon=False)
ax.add_artist(ab)
```
